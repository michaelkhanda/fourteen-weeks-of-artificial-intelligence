```python
#REQUIREMENTS
#Rover carries out two explorations.
#Four different locations
#Rock exploration
#Randomized environmental conditions (wind to cover and uncover rocks periodically)
#Rover should not sample previously sampled location
#Classes: 
#1.	Rover Performance
#2.	Mars Environment
#3.	Rover Agent
#Display final locations
#Rovers performance based on sampling history


#!/usr/bin/env python
# coding: utf-8

# In[12]:


#REQUIREMENTS
#Rover carries out two explorations.
#Four different locations
#Rock exploration
#Randomized environmental conditions (wind to cover and uncover rocks periodically)
#Rover should not sample previously sampled location
#Classes: 
#1.	Rover Performance
#2.	Mars Environment
#3.	Rover Agent
#Display final locations
#Rovers performance based on sampling history


import random,time
class Performance(object):
    def __init__(self):
        self.Score = 0
        self.Visit = 0
    def DisplayScore(self):
        print("Score", self.Score)
        P = (self.Score/self.Visit) * 100
        print("Performance", P, "%")
        return P

class SampleState(object):
    def __init__(self):
        #S = Sample Location, N = Do not Sample Location
        self.State = {"A": "S", "B": "N", "C": "S", "D": "N"}
        self.State["A"] = random.choice(["S", "N"])
        self.State["B"] = random.choice(["S", "N"])
        self.State["C"] = random.choice(["S", "N"])
        self.State["D"] = random.choice(["S", "N"])

class Environment(object):
    def __init__(self):
        self.Location = ["A", "B", "C", "D"]
        #Location State 0 = not sampled, 1 = has already been sampled
        self.LocationState = {"A": "1", "B": "1", "C": "1", "D": "1"}
        self.SampleLocation = random.choice(self.Location)
        self.LocationState["A"] = random.randint(0, 1)
        self.LocationState["B"] = random.randint(0, 1)
        self.LocationState["C"] = random.randint(0, 1)
        self.LocationState["D"] = random.randint(0, 1)
        
class RoverAgent(Environment, SampleState, Performance):
    def __init__(self, Environment, SampleState, Performance):
        print("Current Location: ", Environment.Location)
        print("Location State: ", Environment.LocationState)
        print("State: ", SampleState.State)
        count = 0
        while count < 4:
            Performance.Visit = Performance.Visit + 1
            if Environment.LocationState[Environment.SampleLocation] == 0: #if the rover says the area was not sampled
                Performance.Score = Performance.Score + 1 #add one to the performance
                if SampleState.State[Environment.SampleLocation] == "S": #if the location has not been sampled
                    SampleState.State[Environment.SampleLocation] ==("S") #sample the location
                else:
                    SampleState.State[Environment.SampleLocation] == ("N") #else do not sample the location
                Environment.LocationState[Environment.SampleLocation] = 1 #change to location sampled
                print(Environment.SampleLocation, "Has Been Sampled!") 
            else:
                print(Environment.SampleLocation, " has already been sampled")
            
            newIndex = Environment.Location.index(Environment.SampleLocation) + 1
            if newIndex == 4:
                newIndex = 0
            Environment.SampleLocation = Environment.Location[newIndex]
            count = count + 1
            
        print("Location State: ", Environment.LocationState)
        print("Sample State: ", SampleState.State)
        
TheState = SampleState()
ThePerformance = Performance()
x = 0
TimeInterval = 2
while (x < 6):
    TheEnvironment = Environment()
    TheRover = RoverAgent(TheEnvironment, TheState, ThePerformance)
    
    if x == 2:
        P = ThePerformance.DisplayScore()
        if P > 49:
            print(P, "Adjusting rover tour frequency")
            TimeInterval = 1
            x = 0
        else:
            print(P, "Reducing tour frequency")
            TimeInterval = 3
            x = x + 3
    x = x + 1
    time.sleep(TimeInterval)
ThePerformance.DisplayScore
```
