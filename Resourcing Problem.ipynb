```python
#Objective function: maximize profits = 30000*A + 45000*B
#Decision variables: A,B
#Constraints: A > 0, B > 0, 3A + 4B <= 30, 1.5A + 3B <= 21
import pulp

lp_problem = pulp.LpProblem("Car Manufacturing Example", pulp.LpMaximize)
a = pulp.LpVariable("a", lowBound = 0, cat = "Integer")
b = pulp.LpVariable("b", lowBound = 0, cat = "Integer")

lp_problem += 30000 * a + 45000 * b
lp_problem += 3 * a + 4 * b <= 30
lp_problem += 5 * a + 6 * b <= 60
lp_problem += 1.5 * a + 3 * b <= 21

status = lp_problem.solve()
print(pulp.LpStatus[status])
print(pulp.value(a), pulp.value(b), pulp.value(lp_problem.objective))
```
