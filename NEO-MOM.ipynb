```python
#high blood pressure
#low blood pressure
#high heart rate mother
#high heart rate baby
#low-no heart rate
import random, time
class Performance(object):
    def __init__(self):
        self.Score = 0
        self.Visit = 0
    def DisplayScore(self):
        print("Score", self.Score)
        P = (self.Score/self.Visit) * 100
        print("Performance", P, "%")
        return P
    
class SampleState(object):
    def __init__(self):
        #C = Check on Mother's vital, N = Do not check on Mother's vital
        self.State = {"BP": "C", "HR": "N", "OL": "C"}
        self.State["BP"] = random.choice(["C", "N"])
        self.State["HR"] = random.choice(["C", "N"])
        self.State["OL"] = random.choice(["C", "N"])
        
class Mother(object):
    def __init__(self):
        #0 = normal 1 = abnormal
        #0 = CHECKED 1 = UNCHECKED
        #bp = blood pressure, hr = heart rate, ol = oxygen level
        self.MonitorPoints = ["BP", "HR", "OL"]
        self.MothersState = {"BP": "1", "HR": "1", "OL": "1"}
        self.SampleMother = random.choice(self.MonitorPoints)
        self.MothersState["BP"] = random.randint(0, 1)
        self.MothersState["HR"] = random.randint(0, 1)
        self.MothersState["OL"] = random.randint(0, 1)
        
        
class MonitorAgent(Mother, SampleState, Performance):
    def __init__(self, Mother, SampleState, Performance):
        print("Mother's Vitals: ", Mother.MothersState)
        print("Checked: ", SampleState.State)
        count = 0
        while count < 3:
            Performance.Visit = Performance.Visit + 1
            if Mother.MothersState[Mother.SampleMother] == 0:
                Performance.Score = Performance.Score + 1 
                if SampleState.State[Mother.SampleMother] == "C":
                    SampleState.State[Mother.SampleMother] == ("C")
                else:
                    SampleState.State[Mother.SampleMother] == ("N")
                Mother.MothersState[Mother.SampleMother] = 1
                print(Mother.SampleMother, "Vitals Checked")
            
            else:
                print(Mother.SampleMother, " has already been checked")
                
            newIndex = Mother.MonitorPoints.index(Mother.SampleMother) + 1
            if newIndex == 3:
                newIndex = 0
            Mother.SampleMother = Mother.MonitorPoints[newIndex]
            count = count + 1

        print("Vitals States: ", Mother.MothersState)
        print("Checked: ", SampleState.State , "\n")
        
TheState = SampleState()
ThePerformance = Performance()
x = 0
TimeInterval = 2
while (x < 6):
    TheMother = Mother()
    TheMonitor = MonitorAgent(TheMother, TheState, ThePerformance)
    
    if x == 2:
        P = ThePerformance.DisplayScore()
        if P > 49:
            print(P, "Adjusting vital check frequency")
            TimeInterval = 1
            x = 0
        else:
            print(P, "Reducing vital check frequency")
            TimeInterval = 3
            x = x + 3
    x = x + 1
    time.sleep(TimeInterval)
ThePerformance.DisplayScore
```
